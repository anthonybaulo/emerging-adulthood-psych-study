```python
xl_fname = 'EAMMi2-Data1.2.xlsx'
```

```python
eadf = pd.read_excel(xl_fname)
```

```python
# This will be used below, when grabbing verbose descriptions of columns
label_desc = pd.read_excel(xl_fname, sheet_name='EAMMi2_Labels_Description')
label_desc.head()
```

```python
eadf.tail(3)
```

```python
def col_info(var_name):
    '''Get column description and number of unique values'''
    
    cond = label_desc['Variable Name'] == var_name
    idx = label_desc.index[cond].tolist()[0]
    text = label_desc.iloc[idx]['Question text']
    unique = eadf[var_name].nunique()
    miss = eadf[var_name].isnull().sum()
    
    print(f'Col Name: {var_name}')
    print(f'Desc: {text}')
    print(f'Unique entires: {unique}')
    print(f'Missing: {miss}\n\n')
    
```

```python
# eadf['Q81'].value_counts()
```

```python
cols = ['president','school', 'school_coded', 'affiliation','Q81','Q82','Q83']
for col in cols:
    col_info(col)
```
